```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    #calculating VIF
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
    
    return(vif)
```

```python
#Explanation of variance inflation factor
#In firstline import variance inflation factor from statsmodel
#In secondline Define a function
#In third line Then create a table 
#In fourth line What we give input that is act as a column name that is create as a variable
#In fifth line first execute the oneliner for loop, Shape indicates Number of rows and columns ,
#shape[1] so it take column ,after that no.of values that are go to X.values then we calculate variance inflation factor
#after the calculation its stored in vif
```

## Independent Sample- Unpaired T- Test
### Different Group(male,female) but same condition(salary)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset['gender']=='M']["salary"]
female=dataset[dataset['gender']=='F']["salary"]
#print(MALE)
ttest_ind(male,female)
```

### The P value is 3% so it is good

## Dependant Sample-Paired T-Test

```python
from scipy.stats import ttest_rel
dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']["ssc_p"]
male1 = dataset[dataset['gender']=='M']["hsc_p"]
#print(MALE)
ttest_rel(male,male1)
```

## Two Way Anova

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# two-way ANOVA
anov=ols("mba_p~salary+status",data=dataset).fit()
```

```python
TWA=sm.stats.anova_lm(anov,type=2)
```

```python
#The salary is significant so it accept H1 and reject H0
#The salary is not significant so it accept H0 and reject H1
```
