## Importing the Libraries

```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the Dataset

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
#Number of Rows and Columns
print("The Number of Row:",dataset.shape[0])
print("The Number of Coloumn:",dataset.shape[1])
```

```python
dataset.shape
```

## Univariate Analysis

```python
import seaborn as sb
sb.distplot(dataset["ssc_p"],hist=True,kde=True,kde_kws={"color":"red"})
```

## Bivariate Analysis: Distribution Plot

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset)
plt.show
```

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset,kind="hex")
#adding kind="hex"
```

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset,kind="kde")
plt.show
#adding kind="kde"
```

```python
#Pair Plot
#hue is for differentiate
#diagonal kind: Histogram or kde
#matplot have large number of code but
sb.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl")
plt.show
```

## For Categorical Plot

```python
#strip plot
sb.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

## What is Swarmplot?


```python
# A swarm plot is a type of scatter plot that is used for representing categorical values.
# It is very similar to the strip plot, but it avoids the overlapping of points.
# We can use the seborn.swarmplot() to create such graphs.
#It is not advisable to use this type of graph  wheen the sample size is large.
```

```python
sb.swarmplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

## Distribution Observation


```python
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```

```python
#box plot- means percentile
sb.boxplot(x="ssc_b",y="ssc_p", data=dataset,hue="gender")
plt.show()
```

```python
#Bar plot
sb.barplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender")
plt.show()
```

```python
# Coount plot
sb.countplot(x="ssc_b",data=dataset,hue="gender")
plt.show()
```

```python
# Point Plot
sb.pointplot(x="hsc_p",y="hsc_b",data=dataset,hue="gender")
plt.show()
```

```python
# box plot means percentile
sb.boxplot(data=dataset,orient="h")
plt.show()
```

```python
# factor plot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("exercise")
sb.factorplot(x="time",y="pulse",hue="kind",kind="violin",col="diet",data=df)
plt.show()
```

```python
#factorplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.regplot(x="total_bill",y="tip",data=df)
sb.lmplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
#factorplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("anscombe")
#sb.regplot(x="total_bill",y="tip",data=df)
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"))
plt.show()
```

```python
#factorplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("anscombe")
#sb.regplot(x="total_bill",y="tip",data=df)
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"),order=2)
plt.show()
```

```python
#factorplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
#sb.regplot(x="total_bill",y="tip",data=df)
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```
